```python
# print(plt.style.available)
plt.style.use("seaborn-v0_8-pastel")
```

```python
df1 = df[["Country","W", "D", "L"]]

df1.plot(kind = "bar", x = "Country", stacked= True)
plt.show()
```

```python
tries = df[["Country","TF", "TA"]]

tries.plot(kind = "bar", x = "Country", stacked= True)
plt.show()
```

```python
tries
```

```python
# Création des subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart TF
axes[0].pie(tries['TF'], labels= tries['Country'], autopct='%1.1f%%', startangle=90)
axes[0].set_title("Tries For (TF)")

# Pie chart TA
axes[1].pie(tries['TA'], labels= tries['Country'], autopct='%1.1f%%', startangle=90)
axes[1].set_title("Tries Against (TA)")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
axes = axes.flatten()

labels = ["Tries For", "Tries Against"]
#colors = ["#4CAF50", "#F44336"]

for i, row in tries.iterrows():
    values = [row["TF"], row["TA"]]  # uniquement des nombres
    
    axes[i].pie(
        values,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
    )
    
    axes[i].set_title(row["Country"])

plt.tight_layout()
plt.show()
```
