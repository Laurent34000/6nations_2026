```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://www.espn.com/rugby/standings/_/league/180659/sort/gameswon/dir/desc"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 \
(KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

page = requests.get(url, headers=headers)

print(page.status_code)
```

```python
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

<h3 style = "color: green; text-align: center">Six Nations Standings - 2026</h3>

```python
table = soup.find_all("span", class_ = "fw-medium")
table
```

```python
columns_name = table[1:]

standing_column = [rugby.text.strip() for rugby in columns_name]
print(standing_column)
```

```python
# Countries
countries = soup.find_all("span", class_ = "hide-mobile")

column_countries = [countrie.text.strip() for countrie  in countries]
print(column_countries)
```

```python
# Colonne dans df1
df = pd.DataFrame(columns = column_countries, index = standing_column)
df.head(2)
```

```python
# Column France

france = soup.find_all("span", class_ = "stat-cell")[0:14]

row_france = [rows.text.strip() for rows in france]
print(row_france)

# Column Ireland

ireland = soup.find_all("span", class_ = "stat-cell")[14:28]

row_ireland = [rows.text.strip() for rows in ireland]
print(row_ireland)

# Column Scotland

scotland = soup.find_all("span", class_= 'stat-cell')[28:42]

row_scotland= [rows.text.strip() for rows in scotland]
print(row_scotland)

# Column Italy

italy = soup.find_all("span", class_= 'stat-cell')[42:56]

row_italy = [rows.text.strip() for rows in italy]
print(row_italy)

# Column England
england = soup.find_all("span", class_= 'stat-cell')[56:70]

row_england = [rows.text.strip() for rows in england]
print(row_england)

# Column Walles
wales = soup.find_all("span", class_= 'stat-cell')[70:]

row_wales = [rows.text.strip() for rows in wales]
print(row_wales)
```

```python
# Création des titres des colonnes du DF
#df = pd.DataFrame(columns = rugby_column)

# Remplir le DF et chaque colonne avec les données
df["France"] = row_france
df["Ireland"] = row_ireland
df["Scotland"] = row_scotland
df["Italy"] = row_italy
df["England"] = row_england
df["Wales"] = row_wales

df
```

```python
df = df.transpose()
```

```python
df = df.reset_index()
```
