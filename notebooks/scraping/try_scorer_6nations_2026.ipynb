```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.rugbyrama.fr/resultats/rugby/6-nations/statistiques/joueurs"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('div', class_= "idalgo statistic")[0]
print(table)
```

```python
# Colonne Joueur
players = table.find_all("a", class_ = "player")

row_players = [rows.text.strip() for rows in players]
print(row_players)
```

```python
# Colonne Essais / Match
try_per_match = table.find_all("span", class_= 'value3')[1:]

row_try_per_match = [try_match.text.strip() for try_match in try_per_match]
print(row_try_per_match)
```

```python
# Colonne Match Joués
play = table.find_all("span", class_= 'value2')[1:]
row_play = [game.text.strip() for game in play]
print(row_play)
```

```python
# Colonne Essais
tries = table.find_all("span", class_= 'value1')[1:]

rows_tries = [rows.text.strip() for rows in tries]
print(rows_tries)
```

```python
# Colonne Equipe
nation = table.find_all("a", class_="player-team")

countries = [team.find("span").text.strip() for team in nation]
print(countries)
```

```python
columns_name = table.find_all("span")[0:4]

rugby_column = [rugby.text.strip() for rugby in columns_name]
rugby_column
```

```python
# Création des titres des colonnes du DF
df = pd.DataFrame(columns = rugby_column)

# Remplir le DF et chaque colonne avec les données
df["Marqueurs d'essais"] = row_players
df["essais / match"] = row_try_per_match
df["MJ"] =  row_play
df["essais"] = rows_tries
df["Equipe"] = countries

df.head()
```

```python
df.info()
```

```python
df["MJ"] = pd.to_numeric(df["MJ"])
df["essais / match"] = pd.to_numeric(df["essais / match"].astype(str).str.replace(',', '.'), errors='coerce')
df["essais"] = pd.to_numeric(df["essais"])
```

```python
df = df.rename(columns ={"Marqueurs d'essais": "players","essais / match" : "try_match","essais": "tries", "MJ": "match_played","Equipe": "team" })
df.info()
```

```python
#export en CSV
df.to_csv("../../data/try_scorer.csv", index=False)
```
